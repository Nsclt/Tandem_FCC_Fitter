```julia
include("combined_measurements/inputs.jl")

m_combined = createmodel(observables, measurements_combined, correlations)
m_only_345 = createmodel(observables, measurements_only_345, correlations)
m_only_346 = createmodel(observables, measurements_only_346, correlations)
m_only_347 = createmodel(observables, measurements_only_347, correlations)
m_only_348 = createmodel(observables, measurements_only_348, correlations)
m_only_349 = createmodel(observables, measurements_only_349, correlations)

#CombinationTool.printmodel(m)

likelihood_combined = CombinationDensity(m_combined)
likelihood_only_345 = CombinationDensity(m_only_345)
likelihood_only_346 = CombinationDensity(m_only_346)
likelihood_only_347 = CombinationDensity(m_only_347)
likelihood_only_348 = CombinationDensity(m_only_348)
likelihood_only_349 = CombinationDensity(m_only_349)


prior = NamedTupleDist(
    p1 = [-10.0..10.0],
    p2 = [-20.0..20.0]
)

posterior_combined = PosteriorDensity(likelihood_combined, prior)
posterior_only_345 = PosteriorDensity(likelihood_only_345, prior)
posterior_only_346 = PosteriorDensity(likelihood_only_346, prior)
posterior_only_347 = PosteriorDensity(likelihood_only_347, prior)
posterior_only_348 = PosteriorDensity(likelihood_only_348, prior)
posterior_only_349 = PosteriorDensity(likelihood_only_349, prior)

nchains = 4
nsamples = 10^4
algorithm = MetropolisHastings()

samples_combined, stats_combined = bat_sample(posterior_combined, (nsamples, nchains), algorithm)
samples_only_345 , stats_only_345 = bat_sample(posterior_only_345 , 
(nsamples, nchains), algorithm)
samples_only_346 , stats_only_346 = bat_sample(posterior_only_346 , 
(nsamples, nchains), algorithm)
samples_only_347 , stats_only_347 = bat_sample(posterior_only_347 , 
(nsamples, nchains), algorithm)
samples_only_348 , stats_only_348 = bat_sample(posterior_only_348 , 
(nsamples, nchains), algorithm)
samples_only_349 , stats_only_349 = bat_sample(posterior_only_349 , 
(nsamples, nchains), algorithm)




 function plot_overlay(samples, combination, i, j;
    order=1:length(samples), p=0.9,
    labels=[], seriestypes=[], bins=[], alphas=[], colors=colors)

    color_combination = :deeppink3

    param_labels = ["\${C}_{tZ}\$", "\${C}_{tW}\$"]

    Plots.plot()

    # individual samples
    for k in order
        p = Plots.plot!(
            samples[k], (i,j),
            seriestype=seriestypes[k],
            intervals=[0.9],
            colors=[colors[k]],
            bins=bins[k],
            localmode=false,
            fillalpha = alphas[k],
            label=""
        )
    end

    # combination
    p = Plots.plot!(
        combination, (i,j),
        seriestype=:smallest_intervals_contourf,
        intervals=[0.9],
        colors=[color_combination],
        fillalpha = 0.4,
        bins=300,
        legendfontsize=12,
        legend = :bottomright,
        guidefontsize=22,
        tickfontsize = 18,
        xlims=(-5, 3),
        ylims=(-5, 3),
        xlabel=param_labels[i],
        ylabel=param_labels[j],
        localmode=false,
        label=""
    )


    # SM point
     p = Plots.plot!([0], [0],
        label="SM",
        seriestype = :scatter,
        markershape=:star,
        markerstrokecolor="#595959",#:black,
        markerstrokewidth=0.7,
        markersize=12,
        markercolor=:deeppink)

    # fake legend
    for k in order
        p = Plots.plot!([0], [0],
            label=labels[k],
            seriestype = :shape,
            markershape=:square,
            markersize=10,
            fillcolor=colors[k],
            linewidth=0,
            fillalpha=alphas[k]
        )
    end


    # fake combination legend
    p = Plots.plot!([0], [0],
        label="combination",
        seriestype = :shape,
        markershape=:square,
        markersize=10,
        fillcolor=color_combination,
        linewidth=0,
        dpi = 400
    )

    return p

end


# individual samples
samples_arr = [samples_only_345,samples_only_346,samples_only_347,samples_only_348,samples_only_349]

# combination samples
combination = samples_combined


label1="only 345GeV"
label2="only 346GeV"
label3="only 347GeV"
label4="only 348GeV"
label5="only 349GeV"

labels = [label1, label2, label3, label4, label5]

colors = [:thistle1, :turquoise1, :darkseagreen1, :orchid2, :tomato]
alphas = [0.3, 0.3, 0.3, 0.3, 0.3]

seriestypes = [:smallest_intervals_contour, :smallest_intervals_contour, :smallest_intervals_contour, :smallest_intervals_contour, :smallest_intervals_contour]
bins = [300, 300, 300, 300, 300];




pyplot(size=(600, 500))
p = plot_overlay(samples_arr, combination, 1, 2; order=[1,2,3,4,5], labels = labels,
seriestypes = seriestypes, alphas=alphas, bins=bins, p=0.9)

#plot(samples_combined,xlabel=["\${C}_{tZ}\$" "\${C}_{tZ}\$" "\${C}_{tZ}\$" "\${C}_{tW}\$"],ylabel=["p(\${C}_{tZ}\$)" "\${C}_{tW}\$" "\${C}_{tW}\$" "p(\${C}_{tW}\$)"], legend=[:best :topleft :topleft :best], localmode=false)
#plot!([0], [0],(4,1),
#    label="SM",
#    seriestype = :scatter,
#    markershape=:star,
#    markerstrokecolor="#595959",#:black,
#    markerstrokewidth=0.7,
#    markersize=13,
#    markercolor=:deeppink)



savefig("combined_measurements/Contours_final.pdf")
```
